```python
x=df.iloc[:,:-1] # independent variable
y=df.iloc[:,-1:]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=3)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler
```

```python
# Scaling(Optional)
x_train=scaler.fit_transform(x_train) # for train data use fit_transform
x_test=scaler.transform(x_test)   # for test data use transform
```

```python
x_train
```

```python
x_test
```

```python
# Model Training 
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
x_train.shape
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
# Performance metrics
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
# Adjusted Rsquare
rsq = r2_score(y_test,y_pred)
AdjRsq=1-(1-rsq)*(len(y_test))/(len(y_test)-x_test.shape[1]-1)
AdjRsq
```

```python
# Assumption 
plt.scatter(y_test,y_pred)
plt.show()
```

```python
error = y_test-y_pred
error # error should be normaly distributed
```

```python
sns.distplot(error)
```

```python
# Homoscedasticity >> constatant variance between y_pred and error 
```

```python
plt.scatter(y_pred,error)
plt.show() # it is constant
```
